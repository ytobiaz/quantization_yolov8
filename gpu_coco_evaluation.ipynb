```python
model = 'drive/MyDrive/PA2/models/yolov8x.onnx'
quantized_model = 'drive/MyDrive/PA2/models/yolov8x_quantized.onnx'
```

## YOLOv8

```python
# validate the standard model
args = dict(model=model, data='coco.yaml', device='cuda')
validator = DetectionValidator(args=args)
validator()
```

## Quantized YOLOv8

```python
# validate the quantized model
args = dict(model=quantized_model, data='coco.yaml', device='cuda')
validator = DetectionValidator(args=args)
validator()
```
