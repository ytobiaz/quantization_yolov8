<b>Necessary imports:</b>

```python
import torch
from ultralytics.models.yolo.detect import DetectionValidator
from pycocotools.coco import COCO
```

## YOLOv8

```python
# validate the standard model
args = dict(model='yolov8x.onnx', data='coco.yaml', device='cpu')
validator = DetectionValidator(args=args)
validator()
```

## Quantized YOLOv8

```python
# validate the quantized model
args = dict(model='yolov8x_quantized.onnx', data='coco.yaml', device='cpu')
validator = DetectionValidator(args=args)
validator()
```
