# <b>Visualisation mAP and model size</b>

<b>Necessary imports:</b>

```python
from google.colab import drive
import matplotlib.pyplot as plt
```

```python
# establish connection to google drive
drive.mount('/content/drive')
```

```python
# define data
models = ['YOLOv8 FP32', 'YOLOv8 INT8']
sizes = [260.40, 66.04]  # model sizes in mb
map_scores = [0.537, 0.484]

# create figure
plt.figure(figsize=(8, 6))

# plot the data
plt.plot(sizes, map_scores, 'bo-', label='YOLOv8', markersize=8)

# add markers
for i, (size, map_score) in enumerate(zip(sizes, map_scores)):
    plt.scatter(size, map_score, color='blue', s=100)
    precision_type = 'FP32' if 'FP32' in models[i] else 'INT8'
    plt.text(size, map_score + 0.001, precision_type, fontsize=9, ha='center')

# add labels and legend
plt.xlabel('Model Size (MB)')
plt.ylabel(r'COCO mAP$^{50-95}_{val}$')
plt.legend(loc='lower right')
plt.grid(True)

# save the plot
save_path = '/content/drive/MyDrive/PA2/model_comparison.png'
plt.savefig(save_path)

# show plot
plt.show()
```
